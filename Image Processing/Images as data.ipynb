```python
import cv2 # OpenCV

import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('DL_Data_sets/images/flamingos.jpg')

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(image_gray, cmap='gray')
```

```python
print(image.shape)
print(image_gray.shape)
```

```python
sub_image = image_gray[150:350, 500:750]

plt.imshow(sub_image, cmap='gray')
```

```python
r = image[:, :, 0]
g = image[:, :, 1]
b = image[:, :, 2]

f, (r_ax, g_ax, b_ax) = plt.subplots(1, 3, figsize=(20, 10))

r_ax.set_title('Red channel')
r_ax.imshow(r, cmap='gray')
g_ax.set_title('Green channel')
g_ax.imshow(g, cmap='gray')
b_ax.set_title('Blue channel')
b_ax.imshow(b, cmap='gray')
```

```python
f, (r_ax, g_ax, b_ax) = plt.subplots(1, 3, figsize=(20, 10))

temp_red = np.zeros(image.shape, dtype='uint8')
temp_red[:, :, 0] = r

temp_green = np.zeros(image.shape, dtype='uint8')
temp_green[:, :, 1] = g

temp_blue = np.zeros(image.shape, dtype='uint8')
temp_blue[:, :, 2] = b

r_ax.set_title('Red channel')
r_ax.imshow(temp_red)
g_ax.set_title('Green channel')
g_ax.imshow(temp_green)
b_ax.set_title('Blue channel')
b_ax.imshow(temp_blue)
```
